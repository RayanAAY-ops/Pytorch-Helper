# Pytorch vs Tensorflow

#### Commençons par NumPy :

On peut créer  un algorithme d'apprentissage machine avec NumPy, mais la création d'un réseau de neurones profonds devient de plus en plus difficile. NumPy est conçu pour les calculs (rapides), et non pour l'apprentissage automatique. Il est généralement utilisé pour le machine learning avec un package qui lui est propre.

#### Pytorch:
``PyTorch`` ressemble à ``NumPy`` dans la façon dont il gère les calculs, mais supporte fortement le GPU. Comme NumPy, il a également un backend C (le langage de programmation), donc ils sont beaucoup plus rapides que les bibliothèques Python natives. NumPy pourrait être accéléré par le GPU (avec un peu de code supplémentaire), mais il n’a pas ce puissant support GPU que PyTorch ou TensorFlow offrent. Enfin, PyTorch a été spécialement conçu pour les fonctionnalités GPU en Python.

-moteur de differentiation automatique

#### Tensorflow :

TensorFlow, d’autre part, était écrit principalement en ``C ++`` et`` CUDA`` (langage de programmation de GPU de NVIDIA), et n’a pas été créé spécifiquement pour Python.


#### Comparaison:

PyTorch convient mieux au prototypage rapide en recherche, aux amateurs et aux projets à petite échelle. TensorFlow convient mieux aux déploiements à grande échelle, en particulier lorsque le déploiement multiplate-forme et intégré est pris en compte.

Lorsque vous écrivez du code TensorFlow, il est «compilé» dans un graphique par Python, puis exécuté par le moteur d'exécution TensorFlow. Sur TensorFlow on peine à comprendre cette couche supplémentaire d’indirection. C'est aussi pour cette raison que TensorFlow a quelques concepts supplémentaires à apprendre, tels que ``la session``, ``le graphique``, ``la portée des variables`` et ``les espaces réservés``. De plus, davantage de code passe-partout est nécessaire pour faire fonctionner un modèle de base. Le temps de mise en route nécessaire pour démarrer avec TensorFlow est nettement plus long que PyTorch.

## Graphe Dynamique

>>Les architectures de réseaux neuronaux modernes peuvent avoir des millions de paramètres pouvant être appris. D'un point de vue informatique, la formation d'un réseau de neurones comprend deux phases:

>>Un passage en avant pour calculer la valeur de la fonction de perte.
Une passe en arrière pour calculer les gradients des paramètres pouvant être appris.

>>Le calcul des gradients est au cœur des techniques d’apprentissage automatique les plus modernes. Cela est particulièrement vrai pour les réseaux de neurones, qui utilisent `l'algorithme de rétropropagation` pour mettre à jour les poids. Pour cette raison, Pytorch dispose d’un support puissant et natif pour le calcul par gradient des fonctions et des variables définies dans le cadre. La technique avec laquelle les gradients sont calculés automatiquement pour des calculs arbitraires est appelée différenciation automatique (parfois algorithmique).

>>Les cadres qui utilisent le modèle de graphe de calcul statique implémentent `la différenciation automatique` en analysant le graphe et en y ajoutant des nœuds de calcul supplémentaires qui calculent le gradient d'une valeur par rapport à une autre, reconstituant la règle de chaîne en liant ces nœuds de dégradé supplémentaires aux arêtes. 

>>Cependant, PyTorch n’a pas de graphes de calcul statiques et n’a donc pas le luxe d’ajouter des nœuds de gradient une fois que les autres calculs ont déjà été définis. 

>>Au lieu de cela,


>>PyTorch doit enregistrer ou suivre le flux de valeurs  dans le programme au fur et à mesure qu'elles se produisent, créant ainsi un graphe de calcul de manière dynamique. Une fois ce graphique enregistré, PyTorch dispose des informations nécessaires pour faire reculer ce flux de calcul et calculer les gradients des sorties à partir des entrées.

```python
#grad = gradient value once calculated
#grad_fn =node in the backward graph
#grad_fn point to the operation node
#is_leaf =tensor leaf of the graph ? : x = torch.tensor(dfhskfs)
#not leaf =y =x+1
#when a variable is not a leaf ,we do not calculate the grad for it
#requires_grad = will it be modified ?
```

```python
a=torch.tensor(2.0,requires_grad =True)
b=torch.tensor(3.0,requires_grad=False)
c=a*b
```

lorsque qu'on créait une nouvelle variable en utilisant deux autres natives ,cette dernière hérite de l'attribut 
``requiregrad`` suivant l'union de l'état des deux autres qui ont permis sa création

Une variable qui n'a pas cet attribut ,ne sera pas prise en compte au cours du processus d'accumulation des gradient.


>un graphe de rétropropagation est créé.
>la fonction de multiplication a accès  à une variable de contexte, qui stocke nos variables relatives à l'opération.
 
>Ensuite ,le contexte est passé à  un bloc appelé "backward Operation" , mulbackward hérite 
 
>des attributs de a et b suivant leurs caractéristiques.

>Dans le MulBackward Bloc ,pointé  par la fonction gradient de c , on a une liste  de tuple chacune correspondante  à une opération  qui devra se  faire au niveau des variables natives

```python
#c.backward()
```

>En conclusion ,la dynamicité du graphe nous permet de mieux contrôler ce qui se passe dans notre architecture au plus bas niveau ,ce qui est impossible en utilisant un graphe statique.


> dans PyTorch, chaque niveau de calcul peut être consulté . Un avantage supplémentaire de Pytorch est qu’il nous permet  de mieux comprendre ce qui se passe dans chaque algorithme que nous nous couvrons. Avec une bibliothèque de graphes de calcul statiques telle que TensorFlow, une fois que vous avez exprimé votre calcul de manière déclarative, vous l'envoyez au GPU où il est traité comme une boîte noire. Mais avec une approche dynamique, vous pouvez plonger pleinement dans tous les niveaux du calcul et voir exactement ce qui se passe. ”
